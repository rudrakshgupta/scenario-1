```python
#Average Card Balance(Rupees) 
def A(n):
    if n == 0:
        a = 0
    else:
        a = 1000*((1+0.05)**(n-1))
    return(a)

#Number of customers be x
def X(n):
    if n == 0:
        x = 0
    else:
        x = 1.05**(n-1)
    return(x)

#Interest Rate(Simple Interest, Calculated on daily basis)
r = 0.15

#Membership fee (per card, per month)
def M(n):
    if n == 0:
        m = 0
    else:
        m = 20*X(n)
    return(m)

#Operation Cost (per card, per month)
def O(n):
    if n == 0:
        o = 0
    else:
        o = 25*X(n)
    return(o)


#Affiliation fee (per card, per month)
def F(n):
    if n == 0:
        f = 0
    else:    
        f = 10*X(n)
    return(f)


#Cost of funds in 'n'th month
def C(n):
    if n == 0:
        c = 0
    else:    
        c = (A(n)*X(n) + C(n-1))*0.065/12
        
    return(c) 

#Interest in the 'n'th month
def I(n):
    if n == 0:
        i = 0
    else:    
        i = (A(n)*X(n) + I(n-1))*0.15/3
        
    return(i)    

#Loss Rate
def L(n):
    if n == 0:
        l = 0
    else:
        l = 0.03*(A(n) + I(n))
    return(l)
#Loss rate is 3% of outstanding balance. That is that after every month 3% of the outstanding balance of the customers will be considered defaulted, i.e. the company will suffer a lose of 3% every month on the total outstanding balance.

```

```python
##Question_1 How much profit % is generated in 1 year?
#Solution_1
#Total Cost in the 'n'th month
def TC(n):
    t = C(n) + L(n) + O(n)
    
    return(t)

#Total Revenure in the 'n'th month
def TR(n):
    k = I(n) + M(n) + F(n)
    
    return(k)

#Total cost until 'n'th month
def Total_cost(n):
    if n == 0:
        q = 0 
    else:
        q = TC(n) + Total_cost(n-1)
    
    return(q) 

#Total revenue until 'n'th month
def Total_revenue(n):
    if n == 0:
        y = 0 
    else:
        y = TR(n) + Total_revenue(n-1)
    
    return(y)

#Profit%
def profit_percentage(n):
    if n == 0:
        w = 0 
    else:
        w = (Total_revenue(n) - Total_cost(n))/Total_cost(n) *100
    
    return(w)

#Profit% in an year
print("Profit percentage in an year is ", profit_percentage(12))
```
